# SIS Analysis

```python
import pandas as pd 

sis = pd.read_csv("SIS/sis_summary_results.csv")
sis
```

## SIS BA-200

```python
filtered_sis = sis[["Network", "m", "SteadyState"]]
filtered_sis = filtered_sis[(filtered_sis["Network"] == "BA-200")]
filtered_sis

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(filtered_sis["m"], filtered_sis["SteadyState"], marker='o', linestyle='-')
plt.xlabel("m (Number of New Links per Node)")
plt.ylabel("Steady State Infection Probability")
plt.title("Steady State vs m for BA-200 Network")
plt.grid(True)
plt.show()
```

```python
filtered_sis = sis[["Network", "m", "steady_index:"]]
filtered_sis = filtered_sis[(filtered_sis["Network"] == "BA-200")]
filtered_sis

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(filtered_sis["m"], filtered_sis["steady_index:"], marker='o', linestyle='-')
plt.xlabel("m (Number of New Links per Node)")
plt.ylabel("Time to reach Steady State")
plt.title("steady_index vs m for BA-200 Network")
plt.grid(True)
plt.show()
```

# SIR Analysis

```python
import pandas as pd 

sir = pd.read_csv("SIR/sir_summary_results.csv")
sir
```

## SIR BA-200

```python
filtered_sir = sir[["Network", "m", "extinction_time"]]
filtered_sir = filtered_sir[(filtered_sir["Network"] == "BA-200")]
filtered_sir[["extinction_mean", "extinction_std"]] = filtered_sir["extinction_time"].str.split("+/-", expand=True, regex=False)
filtered_sir["extinction_mean"] = filtered_sir["extinction_mean"].str.strip().astype(float)
filtered_sir["extinction_std"] = filtered_sir["extinction_std"].str.strip().astype(float)

filtered_sir

plt.errorbar(
    filtered_sir["m"],
    filtered_sir["extinction_mean"],
    yerr=filtered_sir["extinction_std"],
    fmt='o-', capsize=5
)
plt.xlabel("m (Number of New Links per Node)")
plt.ylabel("Extinction Time (Mean  Std)")
plt.title("Extinction Time vs m (BA-200 Network)")
plt.grid(True)
plt.show()
```

```python
filtered_sir = sir[["Network", "m", "peak_time"]]
filtered_sir = filtered_sir[(filtered_sir["Network"] == "BA-200")]
filtered_sir[["peak_time_mean", "peak_time_std"]] = filtered_sir["peak_time"].str.split("+/-", expand=True, regex=False)
filtered_sir["peak_time_mean"] = filtered_sir["peak_time_mean"].str.strip().astype(float)
filtered_sir["peak_time_std"] = filtered_sir["peak_time_std"].str.strip().astype(float)

filtered_sir

plt.errorbar(
    filtered_sir["m"],
    filtered_sir["peak_time_mean"],
    yerr=filtered_sir["peak_time_std"],
    fmt='o-', capsize=5
)
plt.xlabel("m (Number of New Links per Node)")
plt.ylabel("peak_time (Mean  Std)")
plt.title("peak_time vs m (BA-200 Network)")
plt.grid(True)
plt.show()
```

```python
filtered_sir = sir[["Network", "m", "peak_size"]]
filtered_sir = filtered_sir[(filtered_sir["Network"] == "BA-200")]
filtered_sir[["peak_size_mean", "peak_size_std"]] = filtered_sir["peak_size"].str.split("+/-", expand=True, regex=False)
filtered_sir["peak_size_mean"] = filtered_sir["peak_size_mean"].str.strip().astype(float)
filtered_sir["peak_size_std"] = filtered_sir["peak_size_std"].str.strip().astype(float)

filtered_sir

plt.errorbar(
    filtered_sir["m"],
    filtered_sir["peak_size_mean"],
    yerr=filtered_sir["peak_size_std"],
    fmt='o-', capsize=5
)
plt.xlabel("m (Number of New Links per Node)")
plt.ylabel("peak_size (Mean  Std)")
plt.title("peak_size vs m (BA-200 Network)")
plt.grid(True)
plt.show()
```

```python
filtered_sir = sir[["Network", "m", "max_R_mean"]]
filtered_sir = filtered_sir[(filtered_sir["Network"] == "BA-200")]
filtered_sir

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(filtered_sir["m"], filtered_sir["max_R_mean"], marker='o', linestyle='-')
plt.xlabel("m (Number of New Links per Node)")
plt.ylabel("Total number of Infected nodes")
plt.title("Total Infection Size vs m for BA-200 Network")
plt.grid(True)
plt.show()
```
